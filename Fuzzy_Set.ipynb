Fuzzy Set

```python
import numpy as np
import matplotlib.pyplot as plt



def fuzzy(gamma):
  arr = np.linspace(0,20,100)
  x_list = []

  for i in arr:
    factor = gamma * abs(((10/max(10,i)) - (i/max(10,i))))

    if(factor == 0):
      x = 1
    if(factor > 0 and factor <= 1):
      x = 1-factor
    if(factor > 1):
      x = 0
    x_list.append(x)

  return arr,x_list


for i in range(1,6):
  arr,x_list = fuzzy(i)
  plt.plot(arr,x_list)
  plt.show()
```
